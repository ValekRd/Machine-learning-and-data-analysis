```python
import numpy as np
import math
%matplotlib inline
from matplotlib import pylab as plt
```

### 1. Приближение фунции многочленом первой степени

```python
A = np.array([[1, 1], [1, 15]])
f1 = np.sin(1.0 / 5) * math.e**(1.0 / 10) + 5 * math.e**(-1.0 / 2)
f2 = np.sin(15.0 / 5) * math.e**(15.0 / 10) + 5 * math.e**(-15.0 / 2)
b = np.array([f1, f2])
```

```python
a = np.linalg.solve(A, b)
```

```python
x = np.arange(0, 16, 0.1)
y = np.sin(x / 5) * math.e**(x / 10) + 5 * math.e**(-x / 2)
plt.plot(x, y)

v = np.arange(0, 16, 0.1)
u = a[0] + a[1] * v
plt.plot(v, u)

plt.show()
```

### 2. Приближение фунции многочленом второй степени

```python
A = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
f1 = np.sin(1.0 / 5) * math.e**(1.0 / 10) + 5 * math.e**(-1.0 / 2)
f2 = np.sin(8.0 / 5) * math.e**(8.0 / 10) + 5 * math.e**(-8.0 / 2)
f3 = np.sin(15.0 / 5) * math.e**(15.0 / 10) + 5 * math.e**(-15.0 / 2)
b = np.array([f1, f2, f3])
```

```python
a = np.linalg.solve(A, b)
```

```python
x = np.arange(0, 16, 0.1)
y = np.sin(x / 5) * math.e**(x / 10) + 5 * math.e**(-x / 2)
plt.plot(x, y)

v = np.arange(0, 16, 0.1)
u = a[0] + a[1] * v + a[2] * (v*v)
plt.plot(v, u)

plt.show()
```

### 3. Приближение фунции многочленом третьей степени

```python
A = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
f1 = np.sin(1.0 / 5) * math.e**(1.0 / 10) + 5 * math.e**(-1.0 / 2)
f2 = np.sin(4.0 / 5) * math.e**(4.0 / 10) + 5 * math.e**(-4.0 / 2)
f3 = np.sin(10.0 / 5) * math.e**(10.0 / 10) + 5 * math.e**(-10.0 / 2)
f4 = np.sin(15.0 / 5) * math.e**(15.0 / 10) + 5 * math.e**(-15.0 / 2)
b = np.array([f1, f2, f3, f4])
```

```python
a = np.linalg.solve(A, b)
```

```python
x = np.arange(0, 16, 0.1)
y = np.sin(x / 5) * math.e**(x / 10) + 5 * math.e**(-x / 2)
plt.plot(x, y)

v = np.arange(0, 16, 0.1)
u = a[0] + a[1] * v + a[2] * (v**2) + a[3] * (v**3)
plt.plot(v, u)

plt.show()
```
