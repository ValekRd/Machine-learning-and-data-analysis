```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

generate a sample of 1000 from a Maxwell distribution (https://www.encyclopediaofmath.org/index.php/Maxwell_distribution)

$f(x) = \sqrt{\frac{2}{\pi}}x^2 e^{-x^2/2}$,    $\sigma = 1$



```python
maxwell_rv = sts.maxwell()
sample = maxwell_rv.rvs(1000)
```

Construct a histogram and theoretical density

```python
plt.hist(sample, bins=20, label='histogram', normed=True)

x = np.linspace(0,5,100)
pdf = maxwell_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Mathematical expectation and variance (https://www.encyclopediaofmath.org/index.php/Maxwell_distribution):

$EX = 2\sqrt{\frac{2}{\pi}}$ Ð¸ $DX = \frac{3\pi - 8}{\pi}$



```python
EX = 2*math.sqrt(2/math.pi)
DX = (3*math.pi-8)/math.pi
```

Take a sample of volume $n = 3, 5, 10,30, 50$, and construct a histogram of the distributions of their sample means, and plot the normal distribution with parameters:

$\mu = EX, \sigma = \sqrt{\frac{DX}{n}}$

```python
for n in 3, 5, 10 ,30, 50:
    sample = list()
    for i in range(1000):
        smpl = maxwell_rv.rvs(n)
        smpl_av = sum(smpl)/len(smpl)
        sample.append(smpl_av)
    plt.hist(sample, bins = 20, label='histogram', normed = True)

    norm_rv = sts.norm(loc=EX, scale=math.sqrt(DX/n))
    x = np.linspace(0,3.5,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='normal distribution')
    plt.legend()
    plt.ylabel('fraction of samples')
    plt.xlabel('$x (n = %d)$' % n)
    plt.show()
   
```

As you can see, the normal distribution matches well with the histogram, with the increase in the sample size (Central limit theorem works)
