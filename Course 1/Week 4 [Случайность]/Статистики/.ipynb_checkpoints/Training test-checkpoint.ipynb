```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
from math import sqrt
```

```python
# зададим нормально распределенную случайную величину
mu = 1197.69
dispersion = 116182.2
sigma = sqrt(dispersion)

norm_rv = sts.norm(loc=mu, scale=sigma)
```

```python
x = np.linspace(0,2500,10000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# распределение среднего количества осадков
n = 30
dispersion = 116182.2 / n
sigma_av = sqrt(dispersion)

norm_av = sts.norm(loc=mu, scale=sigma_av)
```

```python
x = np.linspace(900,1500,10000)
pdf1 = norm_av.pdf(x)
plt.plot(x, pdf1)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
