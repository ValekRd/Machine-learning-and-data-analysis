#  Глобальная оптимизация

```python
import numpy as np
import scipy
from scipy.optimize import minimize
from scipy.optimize import differential_evolution
import math
```

```python
%matplotlib inline
from matplotlib import pylab as plt
```

### Применим к f(x) метод глобальной оптимизации — дифференциальную эволюцию

```python
x = np.arange(1, 30, 0.1)
def func(x):
    return (np.sin(x / 5.) * math.e**(x / 10.) + 5 * math.e**(-x / 2.))
plt.plot(x, func(x))
plt.show()
```

```python
print differential_evolution(func, [(1, 30)])
```
